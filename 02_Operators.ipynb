```python
### Operadores Aritmeticos ###

# operaciones con enteros
print(3+4)
print(3-4)
print(3*4)
print(3/4)
print(10 % 3)
print(10 // 3)
print(2 ** 3)
print(2 ** 3 + 3 -7 / 1 // 4)

# Operaciones con cadenas de texto
print("Hola" + "python" +"¿Que tal?")
prin("Hola" + str(5))

# Operaciones mixtas
print("Hola " * 5)
print("Hola " * (2 ** 3))

my_float = 2.5 * 2
print("Hola " * int(my_float))

### Operadores conparativos ###

# Operaciones con enteros
print(3 > 4)
print(3 < 4)
print(3 >= 4)
print(4 <= 4)
print(3 == 4)
prin(3 != 4)

# Operaciones con cadenas de texto
 print("Hola" > "Python")
print("Hola"< "Python")
print("aaa" >= "abaa")  # Ordenacion alfabetica por ASCII
print(len("aaa") >= len("abaa"))  # cuenta caracteres
print("Hola" <= "Python" )
print("Hola" == "Hola")
print("Hola" != "Python")

### Operadores logicos ###

  # Basada en el Algebra de Boole https://ea.wikipedia.org/wiki/%c3%811gebra de boole
 print(3 > 4 and "Hola" > "Python")
 print(3 > 4 or "Hola" > "Python")
 print(3 < 4 and "Hola" < "Python")
 print(3 < 4 or "Hola" > "Python")
 print(3 < 4 or ("Hola" > "Python" and 4 == 4))
 print(not (3 > 4))
```
